```python
import pandas as pd

# Baca file csv
data = pd.read_csv("valorant-stats.csv")
data.info()
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Encoding kolom kategori
categorical_cols = ['Name', 'Weapon Type', 'Wall Penetration']
encoders = {col: LabelEncoder() for col in categorical_cols}
for col in categorical_cols:
    data[col] = encoders[col].fit_transform(data[col])
```

```python
# Normalisasi fitur numerik
scaler = MinMaxScaler()
numeric_cols = ['Price', 'Fire Rate', 'Magazine Capacity', 'HDMG_0','BDMG_0','LDMG_0','HDMG_1','BDMG_1','LDMG_1','HDMG_2','BDMG_2','LDMG_2']
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])
data.info()
data.head()
```

```python
data.info()
data.head()
```

```python
from sklearn.cluster import KMeans

# Membuat list yang berisi inertia
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(data)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# membuat objek KMeans
km5 = KMeans(n_clusters=5).fit(data)
```

```python
# Menambahkan kolom label pada dataset
data['Labels'] = km5.labels_
```

```python
# membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(6, 3))
sns.scatterplot(x=data['Name'], y=data['HDMG_0'], hue=data['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.xlabel('Name')
plt.ylabel('HDMG_0')
plt.show()
```
